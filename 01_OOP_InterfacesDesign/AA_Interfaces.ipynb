There is no unique recipe when selecting one of both types of using matplotlib. Both interfaces had pros and cons.

But in some way, OOP (Object Oriented Programming) can reach more abilities of this tool: `matplotlib`

### Examples mixing both styles

Plotting graphs, maps, and all, has a lot to do with design. In this first part, I will focus in the elements that conform a plot, and study the possibilities of design with those all elements `matplotlib` has.

```python
# Import pyplot
import matplotlib.pyplot as plt
```

The first two elements we need to define are the **figure** and the **axes**.<br>
**Figure** is the top level container. We can consider stablishing the size of the plot as the first thing to do. By default *figure* plots a no dimension container. `figsize` let us stablish the main container dimensions.<br>
**Axes** is the second main container, where the data is going to be plot inside.<br>
We can play with the common property `facecolor` to differentiate *figure* and *axes*


```python
# using OOP interface
fig = plt.figure(figsize=(16, 9), facecolor='#8ecae6') # the light blue container
ax = plt.axes(facecolor='#ffb703') # the orange container
```

Mixing the interfaces.<br>
`plt.gfc()`gets the current figure. `plt.gca()` the current axes.

```python
# import NumPy (for the next notebooks, I`ll load all the libraries needed for all the project at the begining)
import numpy as np

x = np.linspace(0,1,2)
plt.plot(x,x)
plt.title("thisChart")

ax = plt.gca()
print(ax.title)

ax.plot(x, 1 - x)
ax.set_title('thisBetterChart')
print(ax.title)
```

If we are going to mix OOP with MATLAB it is recommended doing it in this way:

MATLAB ----> OOP

The other way can bring us unneeded problems

```python
# import iris dataset from sklearn and make a boxplot with axes methods and then pyplot functions
import pandas as pd
from sklearn.datasets import load_iris
data = load_iris()['data']
df = pd.DataFrame(data)
df
```

```python
fig, ax = plt.figure(figsize=(16,9), facecolor='#e0e1dd'), plt.axes(facecolor='#fed9b7')

df.plot.box(ax = ax)
ax.yaxis.grid(True)
ax.xaxis.grid(False, color='#ffe6a7', linewidth=0.1)

plt.tight_layout()
plt.savefig('irisbox.pdf')
```
